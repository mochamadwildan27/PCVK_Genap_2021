5. Buat 3 kotak berbagai ukuran dan warna yang berbeda dalam satu image.
disarankan menggunakan bilangan acak/random!

```
import random
blank1 = np.zeros(shape=(512,512,3),dtype=np.int32)
count = 0
while (count < 3):
  k_b = random.randint(0,511)
  k_k = random.randint(0,511)
  kotak2 = cv.rectangle(blank1, pt1=(k_b,k_b), pt2=(k_k,k_k), color=(k_b,k_k,random.randint(0,255)),thickness=0)
  count+=1
  plt.imshow(kotak2)
```
