```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.feature import corner_peaks
from skimage.morphology import (erosion, dilation, closing, opening,
                                area_closing, area_opening)
from skimage.color import rgb2gray
```

```python
# Number 3
img = cv.imread('/content/drive/MyDrive/Images/lena.jpg')

kernel = np.ones((5,5), np.uint8)
dilation = cv.dilate(img, kernel, iterations = 1)

plt.subplot(121),plt.imshow(img),plt.title('Initialize Citra')
plt.subplot(122),plt.imshow(dilation),plt.title('Results from Dilation Citra')
```

```python
# Number 4
kernel = np.ones((5,5), np.uint8)
erosion = cv.morphologyEx(img, cv.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initialize Citra')
plt.subplot(122), plt.imshow(erosion), plt.title('Results from Eruption Citra')
```

```python
#Number 5
kernel = np.ones((5,5), np.uint8)
dilation = cv.morphologyEx(img, cv.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initialize Citra')
plt.subplot(122), plt.imshow(dilation), plt.title('Results from Dilation Citra')
```

```python
#Number 6

kernel = np.ones((5,5), np.uint8)
erosion = cv.morphologyEx(img, cv.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initialize Citra')
plt.subplot(122), plt.imshow(erosion), plt.title('Results from Eruption Citra')
```

```python
#Number 7
kernel = np.ones((7,7), np.uint8)
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initialize Citra')
plt.subplot(122), plt.imshow(opening), plt.title('Results from Opening Citra')
```

```python
#Without Library
element = cv.getStructuringElement(cv.MORPH_RECT,(7,7))
erosion_dst = cv.erode(img, element)
opening_dst = cv.dilate(erosion_dst, element)

plt.subplot(121), plt.imshow(img), plt.title('Initialize Citra')
plt.subplot(122), plt.imshow(opening_dst), plt.title('Results from Opening Citra')
```

```python
#Number 8
kernel = np.ones((7,7), np.uint8)
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initialize Citra')
plt.subplot(122), plt.imshow(closing), plt.title('Results from Closing Citra')
```

```python
#Without Library
element = cv.getStructuringElement(cv.MORPH_RECT,(7,7))
dilate_dst = cv.dilate(img, element)
closing_dst = cv.erode(dilate_dst, element)

plt.subplot(121), plt.imshow(img), plt.title('Initialize Citra')
plt.subplot(122), plt.imshow(closing_dst), plt.title('Results from Closing Citra')
```
